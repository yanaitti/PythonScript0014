# Import of Libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
import copy
import seaborn as sns
import pickle
import time
from tqdm import tqdm
```

# Open the web sites(1999 - 2020)

```python
soup = []

for year in tqdm(range(1999, 2021)):
#     print(year)
    url_name = 'https://ipojp.com/schedule/?y='+str(year)
    url = requests.get(url_name)
    soup.append(BeautifulSoup(url.content, 'html.parser'))
```

```python
all_data = []

for _soup in tqdm(soup):
    _tmp = copy.copy(_soup)
    tables = _tmp.findAll('table')

    for table in tables:
        rows = table.findAll('tr')

        datas = []

        for row in rows:
            csvRow = []
            try:
                for cell in row.findAll(['td', 'th']):
                    csvRow.append(cell.get_text())
                datas.append(csvRow)
            except:
                pass
        all_data.append(datas)
```

```python
all_data[-1]
```

```python
_tmp = []

for data in tqdm(all_data):
    row = {}
    row['name'] = data[0][1].split('\n')[1]
    row['date'] = data[1][1]
    row['code'] = data[1][3]
    row['m_name'] = data[1][5]
    row['m_key'] = data[2][1][1:]
    row['p_share_sum'] = data[2][3][:-1].replace(',','')
    row['c_month'] = data[2][5][:-1]
    row['p_kari'] = data[3][1]
#     row['p_uri'] = data[3][3][:-1].replace(',','')
#     row['hatsune'] = data[3][5][:-1].replace(',','')
#     row['bb'] = data[4][1]
#     row['p_per'] = data[4][3][:-1]
    row['y'] = data[4][5][:-1]
#     row['content'] = data[5][1]
    _tmp.append(row)
```

## File Save & Load

### Save

```python
## 保存
f = open('list.txt', 'wb')
list_row = _tmp
pickle.dump(list_row, f)
```

### Load

```python
## 呼び出し
f = open("list.txt","rb")
all_data = pickle.load(f)
```

```python
soup2 = []

for row2 in tqdm(all_data):
#     row2 = row[1]
    url_name = 'https://ipojp.com/schedule/details/?id=' + str(row2['code']) + '&pubdt=' + row2['date'].replace('/', '')
    url = requests.get(url_name)
    soup2.append(BeautifulSoup(url.content, 'html.parser'))
#     time.sleep(1)
```

```python
all_data2 = []

for _soup in tqdm(soup2):
    _tmp = copy.copy(_soup)
    tables = _tmp.findAll('table')
    _row = []

    for table in tables:
        rows = table.findAll('tr')

        datas = []

        for row in rows:
            csvRow = []
            try:
                for cell in row.findAll(['td', 'th']):
                    csvRow.append(cell.get_text())
                datas.append(csvRow)
            except:
                pass
        _row.append(datas)
    all_data2.append(_row)
```

```python
## 保存
f = open('list2.txt', 'wb')
list_row = all_data2
pickle.dump(list_row, f)
```

```python
## 呼び出し
f = open("list2.txt","rb")
list_row = pickle.load(f)
```

```python
list_row[-1]
```

```python
_tmp2 = []

for _row in tqdm(list_row):
#     print(_row)
    row = {}
    for _data in _row:
        try:
            if _data[1][0] == '業種':
                row['category'] = _data[1][1]
        except:
            pass

        try:
            if _data[0][2] == '売買単位':
                row['s_unit'] = _data[0][3][:-1].replace(',','')
        except:
            pass

        try:
            if _data[1][0] == '公募株数':
                row['p_offer'] = _data[1][1]
        except:
            pass

        try:
            if _data[1][2] == '売出株数':
                if _data[1][3] == '---':
                    row['s_share'] = '0'
                else:
                    row['s_share'] = _data[1][3][:-1].replace(',','')
        except:
            pass

        try:
            if _data[2][0] == 'オーバーアロットメント':
                row['o_allot'] = _data[2][1]
        except:
            pass

        try:
            if _data[3][0] == '発行済み株数':
                _t1 = _data[3][1].split(' / ')
                _t2 = _t1[0].split('：')
                row['i_share'] = _t2[1][:-1].replace(',','')
        except:
            pass

        try:
            if _data[4][0] == '資本金':
                _t1 = _data[4][1].split(' / ')
                _t2 = _t1[0].split('：')
                row['capital'] = _t2[1][:-1].replace(',','')
        except:
            pass
        
    _tmp2.append(row)
# _tmp2

    
    
# 行で見てはダメ。配列に何のデータが入っているかを見る
# また予測データとして取得できるものを使う 

# 業種
# 売買単位
# 公募株数
# 売出株数
# オーバーアロットメント
# ロックアップ
# 発行済み株数
# 資本金
# x 財務データ　※大項目
# 　直近四半期　連結　売上高、経常利益、当期利益、純資産
```

## データ加工

```python
## ゴミデータは削除
all_df = all_df.drop(all_df[all_df['name'] == '株式会社　インターネットイニシアティブ'].index)
all_df = all_df.drop(all_df[all_df['name'] == 'オオクボ'].index)
all_df = all_df.drop(all_df[all_df['name'] == '安田倉庫'].index)
all_df = all_df.drop(all_df[all_df['name'] == '日本オラクル'].index)
all_df = all_df.drop(all_df[all_df['name'] == '杏林製薬'].index)
```

```python
all_df['cancel'] = all_df['name'].apply(lambda x: 1 if x.startswith('【取消】') else 0)
```

```python
all_df.sort_values('date')
```

```python
# all_df = all_df[all_df['koutou'] != '--']
# all_df = all_df[(~all_df['p_uri'].str.contains('2020')) & (all_df['cancel'] == 0)]
all_df = all_df[all_df['cancel'] == 0]
all_df = all_df.dropna(subset=['m_key'])
```

```python
# all_df['v_kobo'] = all_df['v_kobo'].astype('int64')
# all_df['v_tani'] = all_df['v_tani'].astype('int')
# all_df['d_uri'] = all_df['d_uri'].astype('int64')
# all_df['d_stocked'] = all_df['d_stocked'].astype('int64')
# all_df['d_capital'] = all_df['d_capital'].astype('int64')
all_df = all_df[all_df['m_key'] != '']
all_df = all_df.sort_values(by='date', ascending=False).reset_index(drop=True)
all_df = all_df.drop('cancel', axis=1)
```

```python
train = all_df[all_df['y'] != '--']
train['y'] = train['y'].astype('float')

test = all_df[all_df['y'] == '--'].drop('y', axis=1)
```

## Save

```python
train.to_csv('train.csv', index=None)
test.to_csv('test.csv', index=None)
```
